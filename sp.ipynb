```python
import numpy as np
import pandas as pd
from SaintPetersburg import estimation
```

```python
runs = 100000
max_bid = 16
max_trials_exponent = 4
df = estimation(trials_set=np.logspace(1,max_trials_exponent,num=max_trials_exponent), 
                                                    bids=np.array(list(range(1,max_bid))), runs=runs)
```

```python
ax1 = df.plot.scatter(x='trials', y='s_rate', c='bid', colormap='viridis')
```

```python
import matplotlib.pyplot as plt


fig = plt.figure()
ax1 = fig.add_subplot(111)

for i, bid in enumerate(np.unique(df.bid.values)):
    ax1.scatter(df[df['bid']==bid].trials.values, df[df['bid']==bid].s_rate.values, s=10, c=f'C{i}', 
                                                                            marker="s", label=f'bid = {bid}')
    ax1.plot(df[df['bid']==bid].trials.values, df[df['bid']==bid].s_rate.values, c=f'C{i}')
plt.legend(bbox_to_anchor=(1, 1.05), prop={'size': 8})
ax1.set_xscale('log')
ax1.set(ylabel = 'success rate', xlabel = 'number of trials')
plt.savefig('SSP_success_rate.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth = 0.25
fig = plt.subplots()

trials_set = np.flip(np.unique(df.trials.values))
bids = np.unique(df.bid.values)

# Set position of bar on X axis
br1 = np.arange(len(bids))
br2 = [x + barWidth for x in br1]
#br3 = [x + barWidth for x in br2]
 
# Set colors
num_trials = len(trials_set)
green = np.linspace(1, 0.4, num_trials)
red = 0. * green
blue = red
colors_gain = np.array([red, green, blue]).T
colors_loss = np.array([green, red, blue]).T

# set height of bar
for i, trials in enumerate(trials_set):
        gains = [df[(df['bid']==bid) & (df['trials']==trials)].gains_m.values[0] for bid in bids]
        plt.bar(br1, gains, color=colors_gain[i], width=barWidth, edgecolor='grey', 
                                                                        label=f'average gain in {trials} trials')
for i, trials in enumerate(trials_set):
        losses = [-df[(df['bid']==bid) & (df['trials']==trials)].losses_m.values[0] for bid in bids]
        plt.bar(br2, losses, color=colors_loss[i], width=barWidth, edgecolor ='grey', 
                                                                        label =f'average loss in {trials} trials')
 

 
# Adding Xticks
plt.xlabel('bid', fontsize = 15)
plt.ylabel('ammount', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(gains))], bids)

plt.yscale('log')
#plt.xscale('log')
plt.legend(bbox_to_anchor=(1, 1.05), prop={'size': 10})

plt.savefig('SSP_ammount.png', dpi=300, bbox_inches='tight')
plt.show()
```
